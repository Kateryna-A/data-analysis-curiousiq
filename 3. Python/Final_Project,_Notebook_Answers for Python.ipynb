## Replace f/t values of instant bookable column to No/Yes respectively (hint: inplace = True)

```python
#inplace=True
listings['instant_bookable'].replace(to_replace ='f', value = 'No', inplace = True)
listings['instant_bookable'].replace(to_replace ='t', value = 'Yes', inplace = True)
```

```python
listings.sample(3)
```

## Sort dataset by price in descending order

```python
#inplace=True
listings.sort_values('price', ascending=False, inplace= True)
```

## Fill empty (nan) values of host_response_time to N/A

```python
#inplace=True
listings['host_response_time'].fillna("N/A", inplace = True)
```

```python
listings.sample(3)
```

```python
gr = listings.groupby('host_response_time').count()
gr['listing_id']
```

## Find how many listings per host_response_time in a pie chart

```python
labels = list(listings['host_response_time'].unique())
sizes = list(listings['host_response_time'].value_counts())
explode = (0.1, 0, 0, 0, 0)  # only "explode" the 1st slice (The biggest)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

## Find the average price per city and visualize it in a sorted, bar chart

```python
avg_price = listings.groupby('city')['price'].mean().sort_values()
print(avg_price)
```

```python
data = dict(avg_price)
data_sorted = dict(sorted(avg_price.items(), key=lambda x: x[1])) # feel free to search on the Internet :)
```

```python
city = list(data_sorted.keys())
price = list(data_sorted.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(city, price, color ='blue', width = 0.4)
 
#Add label to x axis 
plt.xlabel('City')
#Add label to y axis
plt.ylabel('Average price')
#Add title to plot
plt.title('Average price per city')

plt.show()
```
